```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
data=load_iris()
print("classes :", data.target_names)
```

```python
X=data.data
Y=data.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=47,test_size=0.25) # value of random state indicates the starting point
```

```python
clf=DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(x_train,y_train)
```

```python
y_predict=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy on train data :',accuracy_score(y_train,clf.predict(x_train)))
print('Accuracy on test data :',accuracy_score(y_test,y_predict))
```

### Tuning the Parameters

```python
clf=DecisionTreeClassifier(criterion='entropy',min_samples_split=50) #incresing the value of the decision tree smoothens the decision tree and prevents overfitting
clf.fit(x_train,y_train)
y_predict=clf.predict(x_test)
print('Accuracy on train data :',accuracy_score(y_train,clf.predict(x_train)))
print('Accuracy on test data :',accuracy_score(y_test,y_predict))   
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

clf=DecisionTreeClassifier(criterion='entropy',min_samples_split=50) #incresing the value of the decision tree smoothens the decision tree and prevents overfitting
clf.fit(X,Y)


tree.plot_tree(clf,filled=True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('mobiles.csv')
```

```python
df['Gender'].replace('Female',0, inplace=True)
df['Gender'].replace('Male',1, inplace=True)
df.head()
```

```python
x=df.iloc[:,:3]
y=df.iloc[:,-1]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=47,test_size=0.25)
```

```python
clf=DecisionTreeClassifier(criterion='entropy')
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_test.shape
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy on train data :',accuracy_score(y_train,clf.predict(x_train)))
print('Accuracy on test data :', accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn import tree

clf=DecisionTreeClassifier(criterion='entropy',min_samples_split=50) #incresing the value of the decision tree smoothens the decision tree and prevents overfitting
clf.fit(x_train,y_train)


tree.plot_tree(clf,filled=True)
plt.show()
```

Choose any of the dataset suitable for regression and apply decision tree and compare with linear regression

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
data=load_boston()
```

```python
X=data.data
Y=data.target
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=47,test_size=0.25)
```

```python
clf=DecisionTreeRegressor()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
print('mean_squared_error :',mean_squared_error(y_test,y_pred))
print('r2score :',r2_score(y_test,y_pred))
print('mean_absolute_error :',mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lim=LinearRegression()
lim.fit(x_train,y_train)
```

```python
y_pred=lim.predict(x_test)
```

```python
print('mean_squared_error :',mean_squared_error(y_test,y_pred))
print('r2score :',r2_score(y_test,y_pred))
print('mean_absolute_error :',mean_absolute_error(y_test,y_pred))
```
